```python
LEVEL OF SIGNIFICANCE=0.05
NULL HYPOTHESIS=ALL ZONES GENERATE SAME AMOUNT OF SALES
ALTERNATE HYPOTHESIS=ZONES GENERATE NOT SAME AMOUNT OF SALES
TAKING A ONE WAY ANOVA TEST
DEPENDENT VARIABLE:AMOUNT OF SALES
INDEPENDENT VARIABLES:ZONEA,ZONEB,ZONEC,ZONED
```

```python
from scipy.stats import f_oneway
fvalue, pvalue = stats.f_oneway(df['Zone - A'], df['Zone - B'], df['Zone - C'], df['Zone - D'])
```

```python
print(fvalue,pvalue)
```

```python
if pvalue<0.05:
   print('reject null hypothesis')
else:
    print('accept null hypothesis')
```
